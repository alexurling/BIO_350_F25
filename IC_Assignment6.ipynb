```python
# Import numpy
import numpy as np
```

```python
# Create an algorithm that calculates the population of 3 ages of a bettle population over 50 years
# The initial First year population is 500 and the inital second year population is 500
# On average 1 beetle produces 1 offspring in second year and 5 offspring in third year
# The first year beetles survive 30% to second year and 30% of second year beetles survive to third year
# Create a definition to calculate and return the population of each age group and the total population for each year with

def beetle_population(years, initial_first_year, initial_second_year):

    # Create the transition matrix and initial population vector
    trap_rate = 0.05
    A = np.array([[0,1,5], [0.3,0,0], [0,0.5,0]]) * (1 - trap_rate)
    initial_population_vector = np.array([initial_first_year, initial_second_year, 0])
    
    # Create lists to store the populations of each age group and total population for each year
    first_year_population = []
    second_year_population = []
    third_year_population = []
    total_population = []

    # Initialize current population vector and append the current populations to the lists
    n_t = initial_population_vector
    first_year_population.append(n_t[0])
    second_year_population.append(n_t[1])
    third_year_population.append(n_t[2])
    total_population.append(sum(n_t))

    # Loop through the number of years to calculate the populations
    for year in range(years):
        # Calculate the population for the next year using matrix multiplication
        n_t = np.dot(A, n_t)
        # Append the current populations to the lists
        first_year_population.append(n_t[0])
        second_year_population.append(n_t[1])
        third_year_population.append(n_t[2])
        total_population.append(sum(n_t))
    # Return the populations of each age group and total population for each year
    return first_year_population, second_year_population, third_year_population, total_population
        
```

```python
first_year, second_year, third_year, total = beetle_population(150, 500, 500)
```

```python
#Graphing the results
import matplotlib.pyplot as plt
years = list(range(151))
# Make the y limit 20000

plt.plot(years, first_year, label='First Year Population')
plt.plot(years, second_year, label='Second Year Population')
plt.plot(years, third_year, label='Third Year Population')
plt.plot(years, total, label='Total Population')
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Beetle Population Over Time')
plt.legend()
plt.grid()
plt.show()
```
