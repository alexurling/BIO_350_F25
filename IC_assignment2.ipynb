- Calculating how quickly duckweed will cover a pond



```python
import matplotlib.pyplot as plt
```

```python
def N_exp(N_0, p, time):
    # Exponential growth model
    # Save n_values into a list
    N_values = [N_0]
    # Save initial population
    current_N = N_0
    for _ in range(time):
        # Update population based on exponential growth
        current_N = current_N + p * current_N
        # Round to 3 decimal places
        current_N = round(current_N, 3)
        # Add current population to the list
        N_values.append(current_N)
    return N_values
```

```python
# Run the function with 40 days and p  = 0.4 with an intial population of 0.01
result = N_exp(0.01, 0.4, 40)
print(result)
```

```python
# Plot the results with days on the x-axis and population on the y-axis

days = list(range(41))
# create the plot
plt.plot(days, result, marker='o')
# Add labels and title
plt.xlabel('Days')
plt.ylabel('Duckweed Coverage')
plt.title('Duckweed Coverage Over Time')
# Add a grid
plt.grid()
# Show the plot
plt.show()
```
