## Creating a program/function that can the following questions from Ch. 7 in Handbook of Q. Ecology.
####  - Program that takes input and answers the following questions
####      - Probability we will find 2 eggs in 1st nest we check?
####      - Probability team will get 3 nests in a row with 1 egg?
####      - Average nr. of eggs per nest?



### - Finallly, be able to adjust the input parameters.
####    - Such as how many nests with 1/2/3 eggs
####    - How many nests in a row with 1 egg

```python
# Parameters
nests_with_1_egg = 4
nests_with_2_eggs = 14
nests_with_3_eggs = 2
nests_in_a_row_with_1_egg = 3

# Create a function to calculate probabilities and averages
def calculate_nest_statistics(nests_with_1_egg, nests_with_2_eggs, nests_with_3_eggs, nests_in_a_row_with_1_egg):
    # Probability we will find 2 eggs in 1st nest we check?
    p_2_eggs = nests_with_2_eggs / (nests_with_1_egg + nests_with_2_eggs + nests_with_3_eggs)
    print(f"Probability of finding 2 eggs in the 1st nest: {p_2_eggs:.2f}")
    
    # Probability team will get 3 nests in a row with 1 egg?
    p_3_nests_1_egg = (nests_with_1_egg / (nests_with_1_egg + nests_with_2_eggs + nests_with_3_eggs)) ** nests_in_a_row_with_1_egg
    print(f"Probability of getting 3 nests in a row with 1 egg: {p_3_nests_1_egg:.2f}")
    
    # Average nr. of eggs per nest?
    average_eggs_per_nest = (nests_with_1_egg + 2 * nests_with_2_eggs + 3 * nests_with_3_eggs) / (nests_with_1_egg + nests_with_2_eggs + nests_with_3_eggs)
    print(f"Average number of eggs per nest: {average_eggs_per_nest:.2f}")

# Call the function to print the output
calculate_nest_statistics(nests_with_1_egg, nests_with_2_eggs, nests_with_3_eggs, nests_in_a_row_with_1_egg)

```
