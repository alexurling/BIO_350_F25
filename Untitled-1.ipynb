Chapter 4 - Lokta-Volterra Competition
- Algorithm to calculate 2 seperate species to visualize interspecific competition

```python
import matplotlib.pyplot as plt
```

```python
# Function to calculate the populations of two competing species over time
def NandM_population(N_0, M_0, K_N, K_M, S_N, S_M, d_N, d_M, time):
    N_values = [N_0]
    M_values = [M_0]
    current_N = N_0
    current_M = M_0
    for _ in range(time):
        next_N = current_N + S_N * current_N * (1 - (current_N / K_N) - (current_M / K_M)) - d_N * current_N
        next_M = current_M + S_M * current_M * (1 - (current_M / K_M) - (current_N / K_N)) - d_M * current_M
        next_N = round(next_N, 3)
        next_M = round(next_M, 3)
        N_values.append(next_N)
        M_values.append(next_M)
        current_N = next_N
        current_M = next_M
    return N_values, M_values

# Initial parameters
N_0 = 12      # Initial population of species N
M_0 = 5
K_N = 1600
K_M = 400
S_N = 1
S_M = 1
d_N = 0.4
d_M = 0.6
time = 35


# Calculate populations
result_N, result_M = NandM_population(N_0, M_0, K_N, K_M, S_N, S_M, d_N, d_M, time)

print("Final populations after", time, "years: N =", result_N[-1], ", M =", result_M[-1])

years_list = list(range(len(result_N)))
plt.plot(years_list, result_N, marker='o', label='Species N')
plt.plot(years_list, result_M, marker='s', label='Species M')
plt.xlabel('Years')
plt.ylabel('Plant Population Size')
plt.title('Plant Population Growth with Self-Shading Competition')
plt.legend()
plt.grid()
plt.show()
```

# Q1: Describe figure
The graph shows how the populations of species N and M change over time, taking into account interspecific competition, where each species directly affects the population size of the other. We can see that at year 10, the growth rate of species N begins to decrease, while the total population of Spcies M shows a steady decline. This shows that the competition for shared space intensified with species N gaining the competetive advantage over species M. The slowing growth of species N shows that it's reaching its carrying capacity.
