```python
#Call the SIR_model and give it S0, I0, R0, beta, d_rate, time and save them into 3 variables (s_list, I_list, R_list)

def SIR_model(S0, I0, R0, beta, d_rate, days):
    S_values = [S0]
    I_values = [I0]
    R_values = [R0]
    Current_S = S0
    Current_I = I0
    Current_R = R0
    for day in range(1, days + 1):
        Next_S = Current_S - (beta * Current_S * Current_I)
        Next_I = Current_I + (beta * Current_S * Current_I) - (d_rate * Current_I)
        Next_R = Current_R + (d_rate * Current_I)
        S_values.append(Next_S)
        I_values.append(Next_I)
        R_values.append(Next_R)
        Current_S = Next_S
        Current_I = Next_I
        Current_R = Next_R
    return S_values, I_values, R_values

#Initial Parameters
S0 = 999
I0 = 1
R0 = 0
beta = 0.001
d_rate = 0.5
time = 30

#Plot the results
S_list, I_list, R_list = SIR_model(S0, I0, R0, beta, d_rate, time)

days_list = list(range(len(S_list)))
plt.plot(days_list, S_list, label='Susceptible')
plt.plot(days_list, I_list, label='Infected')
plt.plot(days_list, R_list, label='Removed')
plt.xlabel('Days')
plt.ylabel('Population')
plt.title('SIR Model Over Time')
plt.legend()
plt.grid()
plt.show()



  
```

# Beta Variable - What hapens when it is changed up or down? What are reasons it may change?
When beta increases, the rate at which the population gets infected increases, and the removal rate increases. When beta decreases, these rates will also decrease.
