## Question: How many beetles are found per 1 m² quadrat in a grassland habitat?

## We assume:
### The average beetle density is λ = 8 beetles per quadrat.

### We sample 100 quadrats randomly across the study site.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Parameters
lambda_rate = 8 # expected beetles per quadrat
num_quadrats = 100

# Generate Poisson-distributed data
beetles = np.random.poisson(lambda_rate, size=num_quadrats)

# Create a DataFrame
df = pd.DataFrame({
    'Quadrat': np.arange(1, num_quadrats + 1),
    'Beetle_Count': beetles
})

# Display first few rows of the DataFrame
print(df.head())

# Plot histogram of beetle counts
count_values, bins, patches = plt.hist(
    df['Beetle_Count'],
    bins=range(0, max(beetles)+2),
    density=True,          # converts to probability density
    alpha=0.6,
    color='forestgreen',
    edgecolor='black',
    label='Observed data'
)

x = np.arange(0, max(beetles)+1)
plt.plot(x, poisson.pmf(x, lambda_rate), 'o-', color='darkred', label='Theoretical Poisson PMF')

# Add labels and legend
plt.title('Beetle Counts per Quadrat (Poisson Distribution Example)')
plt.xlabel('Number of Beetles per Quadrat')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
